```python
import nltk
import spacy
from nltk.corpus import treebank

# =========================================================
# INITIALIZATION & DATASET SETUP
# =========================================================
# Uncomment the following line if running for the first time
# nltk.download('treebank')

print("========== DEPENDENCY PARSING & SENTENCE ANALYSIS ==========")

# Load spaCy's pre-trained model which includes a dependency parser
nlp_spacy = spacy.load("en_core_web_sm")

# Load NLTK Treebank tagged corpus and select sample sentences
nltk_sentences = treebank.sents()[:3]
sample_texts = [" ".join(sent) for sent in nltk_sentences]

print(f"Analyzing {len(sample_texts)} sentences from the Treebank corpus.\n")

# =========================================================
# 1. DEPENDENCY PARSING & SYNTACTIC RELATIONSHIP IDENTIFICATION
# =========================================================
print("========== 1. DEPENDENCY PARSING ==========")

for i, text in enumerate(sample_texts):
    print(f"\n--- Sentence {i+1} ---")
    print(f"Original Text: {text}")
    
    # Process text through spaCy's dependency parser
    doc = nlp_spacy(text)
    
    # Identify and print syntactic relationships
    print(f"{'TOKEN':<15} | {'DEPENDENCY':<12} | {'HEAD TOKEN'}")
    print("-" * 45)
    for token in doc:
        print(f"{token.text:<15} | {token.dep_:<12} | {token.head.text}")

# =========================================================
# 2. CONCEPTUAL COMPARISON: DEPENDENCY VS. CONSTITUENCY
# =========================================================
print("\n========== 2. COMPARISON: DEPENDENCY VS. CONSTITUENCY PARSING ==========")
comparison = """
1. Dependency Parsing:
   - Focus: Focuses on binary relationships between words (Head-Dependent).
   - Representation: Graphs showing grammatical functions (e.g., subject, object, modifier).
   - Use Case: Highly effective for languages with free word order; directly extracts semantic relationships.

2. Constituency Parsing (Phrase Structure Grammar):
   - Focus: Focuses on how words group into phrases (Noun Phrases, Verb Phrases).
   - Representation: Hierarchical tree structures (Parse Trees).
   - Use Case: Better for understanding broad structural hierarchies and formal grammar rules defined by CFGs.
"""
print(comparison)

# =========================================================
# 3. VISUALIZATION (Conceptual Tree representation)
# =========================================================
print("========== 3. VISUALIZING PARSE STRUCTURE ==========")
print("Note: spaCy provides internal tree structures for dependency visualization.")
for i, text in enumerate(sample_texts):
    doc = nlp_spacy(text)
    # Simple ASCII-style visual representation of head-dependent links
    print(f"\nStructure for Sentence {i+1}:")
    for token in doc:
        print(f"  {token.head.text} <--({token.dep_})-- {token.text}")

# =========================================================
# 4. ERROR ANALYSIS & LIMITATIONS
# =========================================================
print("\n========== 4. PARSING ERRORS & LIMITATIONS ==========")
analysis = """
Limitations and Common Parsing Errors:
1. Long-Distance Dependencies: Statistical parsers often struggle to link a subject to a verb if the two are separated by a long clause.
2. Attachment Ambiguity: Prepositional phrase attachment (e.g., 'I saw the man with the telescope') remains a classic challenge where the parser may incorrectly assign the modifier to the verb instead of the noun.
3. Out-of-Vocabulary (OOV) Issues: Parsers trained on news corpora (like Treebank) often fail when processing social media or technical jargon that contains unrecognized tokens or non-standard syntax.
4. Structural Complexity: Highly nested or coordinated structures often exceed the performance capability of transition-based parsers.
"""
print(analysis)
```
