```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords, wordnet
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag
import spacy

# ---------------------------------------------------------
# INITIALIZATION & SETUP
# Download NLTK resources (uncomment if running for the first time)
# nltk.download('punkt')
# nltk.download('stopwords')
# nltk.download('wordnet')
# nltk.download('averaged_perceptron_tagger')

# Load the spaCy English model
nlp_spacy = spacy.load("en_core_web_sm")

# Define the sample text corpus
sample_text = "The quick brown foxes are actively jumping over the lazy dogs."
print(f"Original Text: {sample_text}\n")
# ---------------------------------------------------------


# =========================================================
# 1. PREPROCESSING & LINGUISTIC ANALYSIS: NLTK PIPELINE
# =========================================================
print("========== NLTK PIPELINE ==========")

# Tokenization
nltk_tokens = word_tokenize(sample_text)
print(f"Tokens: {nltk_tokens}")

# Stop-word removal
stop_words = set(stopwords.words('english'))
nltk_filtered = [word for word in nltk_tokens if word.lower() not in stop_words]
print(f"Without Stop-words: {nltk_filtered}")

# Stemming and Lemmatization
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

nltk_stems = [stemmer.stem(word) for word in nltk_filtered]
nltk_lemmas = [lemmatizer.lemmatize(word) for word in nltk_filtered]
print(f"Stems: {nltk_stems}")
print(f"Lemmas: {nltk_lemmas}")

# POS Tagging
nltk_pos = pos_tag(nltk_tokens)
print(f"POS Tags: {nltk_pos}\n")


# =========================================================
# 2. PREPROCESSING & LINGUISTIC ANALYSIS: SPACY PIPELINE
# =========================================================
print("========== SPACY PIPELINE ==========")
doc = nlp_spacy(sample_text)

# Tokenization
spacy_tokens = [token.text for token in doc]
print(f"Tokens: {spacy_tokens}")

# Stop-word removal
spacy_filtered = [token.text for token in doc if not token.is_stop and not token.is_punct]
print(f"Without Stop-words: {spacy_filtered}")

# Lemmatization (spaCy does not have a native stemmer, focusing on more accurate lemmatization)
spacy_lemmas = [token.lemma_ for token in doc if not token.is_stop and not token.is_punct]
print(f"Lemmas: {spacy_lemmas}")

# POS Tagging
spacy_pos = [(token.text, token.pos_, token.tag_) for token in doc]
print(f"POS Tags: {spacy_pos}\n")


# =========================================================
# 3. IDENTIFY MORPHOLOGICAL COMPONENTS (ROOT, SUFFIX)
# =========================================================
print("========== MORPHOLOGICAL COMPONENTS ==========")
# Using the spaCy document to identify roots and approximate suffixes
for token in doc:
    if token.text.isalpha():  # Ensure we are only looking at alphabetical words
        # Simple extraction: if the word starts with its lemma, the remainder is likely the suffix
        suffix = token.text.replace(token.lemma_, "") if token.text.startswith(token.lemma_) and token.text != token.lemma_ else "None"
        print(f"Word: {token.text:10} | Root: {token.lemma_:10} | Suffix: {suffix}")


# =========================================================
# 4. INTERPRET AMBIGUITY (LEXICAL AND SYNTACTIC)
# =========================================================
print("\n========== INTERPRETING AMBIGUITY ==========")

# Sentence A: Lexical Ambiguity
sentence_lexical = "I went to the bank to deposit my money."
print(f"Sentence A: '{sentence_lexical}'")

lexical_tokens = word_tokenize(sentence_lexical.lower())
ambiguous_lexical_word = "bank"

if ambiguous_lexical_word in lexical_tokens:
    print("Classification: Lexical Ambiguity detected.")
    print(f"Reason: The word '{ambiguous_lexical_word}' has multiple distinct dictionary meanings.")
    
    # Retrieve and print definitions from WordNet to prove lexical ambiguity
    synsets = wordnet.synsets(ambiguous_lexical_word)
    if synsets:
        print(f"WordNet Definitions for '{ambiguous_lexical_word}':")
        for i, syn in enumerate(synsets[:2]):
            print(f"  - {syn.definition()}")

print("-" * 40)

# Sentence B: Syntactic Ambiguity
sentence_syntactic = "I saw the man with the telescope."
print(f"Sentence B: '{sentence_syntactic}'")

ambiguous_syntactic_phrase = "with the telescope"

if ambiguous_syntactic_phrase in sentence_syntactic.lower():
    print("Classification: Syntactic Ambiguity detected.")
    print(f"Reason: The phrase '{ambiguous_syntactic_phrase}' creates a structure that can be parsed in multiple valid ways.")
    print("Explanation: It is structurally unclear whether 'with the telescope' modifies the noun 'the man' (the man had a telescope) or the verb 'saw' (the observer used the telescope).")
```
