```python
import string
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords, words
from nltk.stem import PorterStemmer, WordNetLemmatizer

# =========================================================
# INITIALIZATION & SETUP
# =========================================================
# Uncomment the following lines if running for the very first time
# nltk.download('punkt')
# nltk.download('stopwords')
# nltk.download('wordnet')
# nltk.download('words')

print("========== TEXT PREPROCESSING & OOV HANDLING ==========")

# 1. Dataset Initialization
# Simulating a noisy text from a Twitter dataset or IMDB Movie Review
raw_text = "OMG!! The cinemmatography in this MoviE was abso-lutely fantastiiic!!! I loved it. However, the plot was kinda meh and predictable."
print(f"Original Text:\n{raw_text}\n")


# =========================================================
# 2. SENTENCE SEGMENTATION & TOKENIZATION
# =========================================================
print("========== PIPELINE EXECUTION ==========")

# Sentence Segmentation
sentences = sent_tokenize(raw_text)
print("1. Sentence Segmentation:")
print(sentences)
print()

# Tokenization
raw_tokens = []
for sent in sentences:
    raw_tokens.extend(word_tokenize(sent))
print("2. Tokenization:")
print(raw_tokens)
print()


# =========================================================
# 3. TEXT NORMALIZATION, TRUE CASING & STOPWORD REMOVAL
# =========================================================
# Normalization: Lowercasing (a basic form of true casing for general text) and removing punctuation
normalized_tokens = [token.lower() for token in raw_tokens if token not in string.punctuation]

print("3. Text Normalization (Lowercasing & Punctuation Removal):")
print(normalized_tokens)
print()

# Stopword Removal
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in normalized_tokens if word not in stop_words]

print("4. Stopword Removal:")
print(filtered_tokens)
print()


# =========================================================
# 4. STEMMING & LEMMATIZATION
# =========================================================
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

stemmed_tokens = [stemmer.stem(word) for word in filtered_tokens]
lemmatized_tokens = [lemmatizer.lemmatize(word) for word in filtered_tokens]

print("5. Stemming (Porter Stemmer):")
print(stemmed_tokens)
print()
print("6. Lemmatization (WordNet):")
print(lemmatized_tokens)
print()


# =========================================================
# 5. HANDLING OUT-OF-VOCABULARY (OOV) WORDS
# =========================================================
# Simulating a known training vocabulary using NLTK's built-in words corpus
# Adding some context-specific words that a model would typically know
known_vocab = set(words.words()).union({'movie', 'cinematography', 'predictable', 'loved', 'plot', 'however'})

processed_tokens = []
oov_words = []

for token in lemmatized_tokens:
    if token in known_vocab:
        processed_tokens.append(token)
    else:
        # Unknown token replacement
        processed_tokens.append("<UNK>")
        oov_words.append(token)

print("7. OOV Handling (Unknown Token Replacement):")
print(f"Final Tokens: {processed_tokens}")
print(f"Identified OOV Words (misspellings, slang, etc.): {oov_words}")
print()


# =========================================================
# 6. ANALYZE AND COMPARE
# =========================================================
print("========== ANALYSIS & COMPARISON ==========")
print(f"Original Text length (characters): {len(raw_text)}")
print(f"Original Word count (approx)     : {len(raw_text.split())}")
print(f"Final Processed Token count      : {len(processed_tokens)}")
print("\nComparison Summary:")
print("- Noise such as exclamation points and capitalization irregularities (e.g., 'MoviE') were successfully normalized.")
print("- High-frequency grammar words (e.g., 'the', 'was', 'in') were removed to reduce dimensionality.")
print("- Spelling errors and slang ('cinemmatography', 'fantastiiic', 'meh', 'abso-lutely') were successfully identified as OOV and replaced with the <UNK> tag, preventing vocabulary explosion in downstream NLP models.")
```
