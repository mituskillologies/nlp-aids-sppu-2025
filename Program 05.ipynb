```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

print("========== FEATURE EXTRACTION & TEXT REPRESENTATION ==========")

# =========================================================
# 1. DATASET INITIALIZATION
# =========================================================
# Creating a custom dataset of 10 short documents as specified in the syllabus.
documents = [
    "Machine learning is fascinating.",
    "Artificial intelligence and machine learning are related.",
    "Natural language processing extracts features from text.",
    "Text representation is a core NLP task.",
    "Bag of words and TF-IDF are feature extraction techniques.",
    "Sparse vectors have many zeros.",
    "Dense embeddings represent text effectively.",
    "TF-IDF stands for Term Frequency Inverse Document Frequency.",
    "One-hot encoding is a very sparse representation.",
    "NLP converts text into numerical features."
]

print(f"Loaded a custom dataset of {len(documents)} documents.\n")


# =========================================================
# 2. ONE-HOT ENCODING (Binary Document-Term Matrix)
# =========================================================
print("========== 1. ONE-HOT ENCODING ==========")
# Using CountVectorizer with binary=True simulates document-level one-hot encoding
onehot_vectorizer = CountVectorizer(binary=True)
onehot_matrix = onehot_vectorizer.fit_transform(documents)
vocab = onehot_vectorizer.get_feature_names_out()

onehot_df = pd.DataFrame(onehot_matrix.toarray(), columns=vocab)
print("One-Hot Encoded Features (First 3 Documents, First 10 Features):")
print(onehot_df.iloc[:3, :10])
print()


# =========================================================
# 3. BAG OF WORDS (BoW) / DOCUMENT-TERM MATRIX (DTM)
# =========================================================
print("========== 2. BAG OF WORDS (BoW) & DTM ==========")
# Standard CountVectorizer creates a Document-Term Matrix with word frequencies
bow_vectorizer = CountVectorizer(binary=False)
bow_matrix = bow_vectorizer.fit_transform(documents)

bow_df = pd.DataFrame(bow_matrix.toarray(), columns=vocab)
print("Document-Term Matrix (BoW) (First 3 Documents, First 10 Features):")
print(bow_df.iloc[:3, :10])
print()


# =========================================================
# 4. TF-IDF REPRESENTATION
# =========================================================
print("========== 3. TF-IDF ==========")
# TfidfVectorizer calculates the Term Frequency-Inverse Document Frequency
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(documents)

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vocab)
print("TF-IDF Matrix (First 3 Documents, First 10 Features):")
# Rounding for visual clarity
print(tfidf_df.iloc[:3, :10].round(3))
print()


# =========================================================
# 5. FEATURE DIMENSIONALITY & SPARSITY ANALYSIS
# =========================================================
print("========== 4. ANALYSIS: DIMENSIONALITY & SPARSITY ==========")

# Dimensionality
num_documents, num_features = bow_matrix.shape
print(f"Feature Dimensionality: {num_features} unique words across {num_documents} documents.")

# Sparsity Calculation (Percentage of zeros in the matrix)
total_elements = num_documents * num_features
non_zero_elements = bow_matrix.nnz
zero_elements = total_elements - non_zero_elements
sparsity_percentage = (zero_elements / total_elements) * 100

print(f"Total Elements in Matrix : {total_elements}")
print(f"Non-Zero Elements        : {non_zero_elements}")
print(f"Zero Elements            : {zero_elements}")
print(f"Matrix Sparsity          : {sparsity_percentage:.2f}%")
print()


# =========================================================
# 6. COMPARISON: SPARSE VS. DENSE REPRESENTATIONS
# =========================================================
print("========== 5. COMPARISON & INTERPRETATION ==========")
interpretation = f"""
1. Sparse vs. Dense Representations:
   - The generated Document-Term Matrix (DTM) and TF-IDF matrices are highly SPARSE ({sparsity_percentage:.2f}% zeros).
   - Sparse representations (like One-Hot, BoW, and TF-IDF) suffer from the "curse of dimensionality" as the vocabulary grows. They do not capture semantic relationships (e.g., "machine" and "intelligence" are orthogonal/independent features).
   - In contrast, DENSE representations (like Word2Vec or GloVe) would map these words into a lower-dimensional, dense space (e.g., 300 dimensions with mostly non-zero values) where similar words have similar vectors.

2. Feature Value Interpretation:
   - One-Hot: Indicates only the presence (1) or absence (0) of a word in a document.
   - Bag of Words (DTM): Indicates the raw frequency count of a word in a document.
   - TF-IDF: Balances local frequency with global rarity. High values indicate that a word is highly relevant to a specific document, penalizing common words across the corpus.
"""
print(interpretation)
```
